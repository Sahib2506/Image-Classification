```python
import pandas as pd
import numpy as np
from selenium import webdriver
driver = webdriver.Chrome(r"C:\Users\OWNER\Downloads\chromedriver.exe")
```

```python
import time
from selenium.common.exceptions import NoSuchElementException
driver.get('https://amazon.com')
```

```python
driver.maximize_window()
```

```python
#Getting the search bar

images = driver.find_element(by = 'xpath', value= "//input[@class='nav-input nav-progressive-attribute']").send_keys(input("Looking for product : "))

button = driver.find_element(by = 'xpath', value= "//span[@class = 'nav-search-submit-text nav-sprite nav-progressive-attribute']").click()
```

```python
#Getting into only the sarees section

driver.find_element(by = 'xpath', value = "/html/body/div[1]/div[2]/div[1]/div[2]/div/div[3]/span/div[1]/span/div/div/div[1]/ul/li[5]/span/a").click()
```

```python
#Locate the images to be scraped from the current page
link = []

for i in range(1,5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") 
    driver.execute_script("window.scrollTo(0, 0)")
    time.sleep(5)  # sleep_between_interactions
    imgResults = driver.find_elements(by='xpath', value = "//img[@class='s-image']")
    for i in imgResults:
        link.append(i.get_attribute('src'))
    time.sleep(2)
    driver.find_element(by='xpath', value = "//li[@class = 'a-last']").click()
len(imgResults)
```

```python
from PIL import Image
import requests
import io
```

```python
for i, url in enumerate(link):
    file_name = f"{i:15}.jpg"    
    try:
        image_content = requests.get(url).content

    except Exception as e:
        print(f"ERROR - COULD NOT DOWNLOAD {(url)} - {e}")
        
        
    try:
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert('RGB')
        
        file_path = os.path.join(direct, file_name)
        
        with open(file_path, 'wb') as f:
            image.save(f, "JPEG", quality=85)
            print(f"SAVED - {url} - AT: {file_path}")
    except Exception as e:
            print(f"ERROR - COULD NOT SAVE {url} - {e}")
```
